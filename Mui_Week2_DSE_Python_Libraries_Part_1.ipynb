**เขียนโปรแกรมเพื่อวิเคราะห์เบอร์โทรมงคล**

หมายเหตุ: เป็นตัวอย่างเพื่อสำหรับเรียนรู้เรื่องเขียนโปรแกรมเท่านั้น ข้อกำหนดเรื่องเบอร์มงคลในตัวอย่างนี้ไม่สามารถใช้ได้จริงแต่อย่างใด

ข้อกำหนด: เราคำนวณคะแนนเบอร์มงคล ด้วย 3 ขั้นตอน
1. คำนวณคะแนนจากผลรวมของเลขในเบอร์โทร
หากเบอร์ดีจะมีผลรวมของเลขทั้งหมดเท่ากับ 9, 14, 15, 19, 23, 24, 32, 36, 40, 41, 42, 44, 45, 46, 50, 51, 54, 55, 56, 59, 60, 63, 64, 65 (ให้ 20 คะแนน)
2. คำนวณคะแนนจากเลขข้างในทีละตัว
หากเป็นเลขดี 2, 5, 9 (ให้ 3 คะแนน)
หากเป็นเลขไม่ดี 3, 6, 8 (ติดลบ 1 คะแนน)
3. รวมคะแนนจาก 1 และ 2 แล้วเทียบคะแนนกับเกณฑ์
หากได้ผลรวมไม่เกิน 15 เป็นเป็นเฉยๆ
หากได้ผลรวมตั้งแต่ 15 ถึง 25 เป็นเบอร์ดี
หากได้ผลรวมตั้งแต่ 26 ขึ้นไป เป็นเบอร์ดีมาก

**โจทย์ 1: เขียนโปรแกรมเพื่อวิเคราะห์ระดับความดีของเบอร์โทรที่ป้อนเข้ามา**

````python
#รับ input เบอร์โทรจาก user แล้วเก็บไว้ในตัวแปร
phone_no = '08999945548' #@param {type:"string"}
#กำหนดค่าตัวแปรตั้งต้นต่างๆ 
score = 0
good_sum = [9, 14, 15, 19, 23, 24, 32, 36, 40, 41, 42, 44, 45, 46, 50, 51, 54, 55, 56, 59, 60, 63, 64, 65]
good_number = [0,2,5,9]
bad_number = [3,6,8]
phone_total_sum = 0

#คำนวนคะแนนขั้นตอนที่ 1 จากผลรวม​ โดยดึงเลขออกมาทีละตัวเพื่อบวกเข้าไปในผลลัพธ์
for number in phone_no:
  phone_total_sum += int(number)
#นำผลรวมที่ได้ ไปเทียบว่าเป็นผลรวมที่ดีหรือไม่ กำหนดะแนนที่ได้สำหรับขั้นตอนแรก
if phone_total_sum in good_sum:
  score = 20

#คำนวณคะแนนขั้นตอนที่ 2 โดยดูจากแต่ละเลขในเบอร์โทร ให้ดึงเลขออกมาทีละตัว 
#แล้วดูว่าเป็นเลขดีหรือไม่ดี แล้วนำไปบวกหรือลบคะแนนรวม
for number in phone_no:
  if int(number) in good_number:
    score +=3
  elif int(number) in bad_number:
    score -=1

#นำคะแนนรวมทั้งหมดมาเทียบกับเกณฑ์ แล้วแสดงผลความเป็นมงคลของเบอร์
if score <= 15:
  print("ผลวิเคราะห์ของเบอร์ " + phone_no + " : ความเป็นมงคลกลางๆ")
elif score > 15 and score < 26:
  print("ผลวิเคราะห์ของเบอร์ " + phone_no + " : ความเป็นมงคลดี")
else:
  print("ผลวิเคราะห์ของเบอร์ " + phone_no + " : ความเป็นมงคลดีมากๆ")
````

**โจทย์ 2: เขียนโปรแกรมเพื่อเปรียบเทียบคะแนนความเป็นมงคลระหว่างเบอร์ 2 เบอร์**

````python
#ทำโจทย์ที่ 1 ให้เป็น function คำนวณคะแนนรวม
def tel_score(tel):
  my_score = 0
  good_sum = [9, 14, 15, 19, 23, 24, 32, 36, 40, 41, 42, 44, 45, 46, 50, 51, 54, 55, 56, 59, 60, 63, 64, 65]
  good_number = [2,5,9]
  bad_number = [3,6,8]
  phone_total_sum = 0
  
  for number in tel:
    phone_total_sum += int(number)

  if phone_total_sum in good_sum:
    my_score = 20

  for number in tel:
    if int(number) in good_number:
      my_score +=3
    elif int(number) in bad_number:
      my_score -=1

  return my_score
#รับ input เบอร์โทรที่ 1
phone_no_1 = '0896541878' #@param {type:"string"}
#รับ input เบอร์โทรที่ 2
phone_no_2 = '0257489858' #@param {type:"string"}

#โยนเบอร์โทร 1 เข้าไป function คำนวณคะแนนรวม เก็บค่าคะแนนรวมไว้ในตัวแปร 1
score_phone_1 = tel_score(phone_no_1)

#โยนเบอร์โทร 2 เข้าไป function คำนวณคะแนนรวม เก็บค่าคะแนนรวมไว้ในตัวแปร 2
score_phone_2 = tel_score(phone_no_2)

#เปรียบเทียบคะแนนของเบอร์โทรที่ 1 กับเบอร์โทรที่ 2
#แสดงผลว่าเบอร์โทรใดมีคะแนนมากกว่า
if score_phone_1 > score_phone_2:
  print('เบอร์ ', phone_no_1, ' ดีกว่า')
elif score_phone_2 > score_phone_1:
  print('เบอร์ ', phone_no_2, ' ดีกว่า')
else:
  print('เบอร์ทั้งสองดีพอๆ กัน')
````

````python
class car:
  def __init__(self,brand,horsePower,color):
    self.brand = brand
    self.horsePower = horsePower
    self.color = color
    self.currentPosition = 0
 
  def start(self):
    print("start the car")
  def moveForward(self):
    self.currentPosition = self.currentPosition + self.horsePower
  def moveBackward(self):
    self.currentPosition = self.currentPosition - self.horsePower
````
