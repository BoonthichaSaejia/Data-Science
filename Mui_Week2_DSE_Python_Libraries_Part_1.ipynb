Boolean array indexing: Boolean array indexing lets you pick out arbitrary elements of an array. Frequently this type of indexing is used to select the elements of an array that satisfy some condition. Here is an example:

````python
a = np.array([[1,2], [3,4], [5,6]])
print(a)
````

````python
# Find the elements of a that are bigger than 2;
# this returns a numpy array of Booleans of the same
# shape as a, where each slot of bool_idx tells
# whether that element of a is > 2.
bool_idx = (a > 2)
print(bool_idx)
````

````python
# We use boolean array indexing to construct a rank 1 array
# consisting of the elements of a corresponding to the True values
# of bool_idx
print(a[bool_idx])
````

````python
# We can do all of the above in a single concise statement:
print(a[a>2])
````

###Array math

Basic mathematical functions operate elementwise on arrays, and are available both as operator overloads and as functions in the numpy module:

````python
x = np.array([[3,3],[3,3]], dtype=np.float64)
print(x)
````

````python
y = np.array([[2,2],[2,2]], dtype=np.float64)
print(y)
````

````python
# Elementwise sum
print(x + y)
````

````python
# Elementwise sum
print(np.add(x, y))
````

````python
# Elementwise difference
print(x-y)
````

````python
# Elementwise difference
print(np.subtract(x, y))
````

````python
# Elementwise product
print(x * y)
````

````python
# Elementwise product
print(np.multiply(x, y))
````

````python
# Elementwise division
print(x/y)
````

````python
# Elementwise division
print(np.divide(x, y))
````

````python
v = np.array([1,2])
w = np.array([10, 20])
# (1*10)+(2*20)
# Inner product of vectors; both produce 
print(np.dot(v, w))
print(v.dot(w))
````

You can also use the `@` operator which is equivalent to numpy's `dot` operator.

Object

````python
class car:
  def __init__(self,brand,horsePower,color):
    self.brand = brand
    self.horsePower = horsePower
    self.color = color
    self.currentPosition = 0
 
  def start(self):
    print("start the car")
  def moveForward(self):
    self.currentPosition = self.currentPosition + self.horsePower
  def moveBackward(self):
    self.currentPosition = self.currentPosition - self.horsePower
````

````python
car_ricky = car("honda",200,"red")
car_tao = car("toyota",300,"blue")
````

````python
print(car_tao.currentPosition)
car_tao.start()
car_tao.moveForward()
print(car_tao.currentPosition)
````
