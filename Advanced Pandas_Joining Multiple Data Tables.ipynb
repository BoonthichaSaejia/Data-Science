# คำสั่งที่ควรรู้

pd.merge(df1,df2) = join 2 dataframes 
    
    type default = 'inner'
    
    
pd.merge(df1,df2, how='left') = join แบบ df1 สำคัญกว่า 

pd.merge(df1,df2, how='right') = join แบบ df2 สำคัญกว่า 


pd.merge(df1,df2, on="playerID") = เชื่อมกันด้วย index "playerID"

pd.merge(df1,df2, left_on="playerID,right_on="plrID") = ใช้กรณี column ที่ใช้เชื่อมใช้คนละชื่อ

pd.merge(df1,df2,left_index=True, right_on="plrID" = ใช้ index ซ้ายเป็นหลัก

```python
#check that default of how is right
print(
    pd.merge(master,scoring,left_index=True,right_on="playerID").shape,
    pd.merge(master,scoring,left_index=True,right_on="playerID",how="right").shape
)
```

```python
#check that default of how is right

#drop random records
master2=master.drop(master.sample(5).index)
print(
    pd.merge(master2,scoring,left_index=True,right_on="playerID").shape,
    pd.merge(master2,scoring,left_index=True,right_on="playerID",how="right").shape
)
```
