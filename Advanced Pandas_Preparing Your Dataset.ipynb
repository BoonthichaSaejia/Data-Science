#### Df=Df.dropna(subset=['column_name'])

```python
df_master_orig=df_master.copy()
df_master=df_master.dropna(subset=['playerID'])
df_master.shape
```

```python
df_master=df_master.dropna(subset=["firstNHL","lastNHL"], how="all")
#how="all" จะทำให้เราลบrowที่ไม่มีข้อมูลในช่อง firstNHL กับ lastNHL ทั้งคู่
```

```python
df_master.shape
```

```python
#choose row that have lastNHL(เล่นเกมสุดท้ายเมื่อปี 1980) more than 1980
df_master=df_master.loc[df_master["lastNHL"]>=1980]
df_master.shape
```

```python
df_master.head()[["lastNHL"]]
```

```python
columns_to_keep=["playerID","firstName","lastName","pos","birthYear","birthMon","birthDay",
                 "birthCountry","birthState","birthCity"]
df_master[columns_to_keep].head()
```

```python
df_master.filter(columns_to_keep).head()
```

Df.filter(set_of_columns).head()=Df[set_of_columns].head()

```python
master=df_master.filter(regex="(playerID|pos|^birth)|(Name$)")
#^birth=something ขึ้นต้นด้วย birth
#Name$=something ลงท้ายด้วย Name
```

```python
master.columns
```

#### Function check memory

```python
def mem_mib(df):
    print("{0:.2f} MiB".format(
        df.memory_usage().sum()/(1024*1024)
    ))
mem_mib(df_master_orig)
mem_mib(master)
```

```python
master["pos"].value_counts()
```

```python
pd.Categorical(master["pos"])
```

```python
len(pd.Categorical(master["pos"]))
```

#### Function เอามาแค่ทีละ catagory

```python
def make_categorical(df,col_name):
    df.loc[:,col_name]=pd.Categorical(df[col_name])
```

```python
make_categorical(master,"pos")
make_categorical(master,"birthCountry")
make_categorical(master,"birthState")
mem_mib(master)
```

```python
master.head()
```

#### Df.set_index() = เปลี่ยน index 

```python
master=master.set_index("playerID")
master.head()
```

```python
master.to_pickle(os.path.join("..","master.pickle"))
```

#### Function เลือกเฉพาะ row ที่แข่งใน NHL และยังแข่งหลังจากปี 1980

```python
def recent_nhl_only(df):
    return df[(df["lgID"]=="NHL")&(df["year"]>=1980)]

scoring=recent_nhl_only(df_scoring)
scoring.shape
```

```python
scoring_2=scoring.copy
```

```python
scoring=scoring.filter(regex="^(?!(Post|PP|SH)).*")
#^(?!(Post|PP|SH)) ไม่เอาขึ้นต้นด้วย Post,PP,SH
#regex="(playerID|pos|^birth)|(Name$)"
#^birth=something ขึ้นต้นด้วย birth
#Name$=something ลงท้ายด้วย Name
```
